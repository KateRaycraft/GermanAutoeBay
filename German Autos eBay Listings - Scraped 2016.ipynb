The seller, offer_type, and num_of_pics columns mostly have the same value, and can be removed.  The most frequent price is $0, which is interesting.

At first I found it funny that limousine accounted for >25% of the entries, until I discovered that it refers to what we would call a sedan in North America. Additionally, 25% of the vehicles listed look like they're from before 1999 - which means they're at least 17 years old at the time of listing, which is geriatric for a vehicle.  There's also at least one vehicle listed from the year 9999, and one from 1000 - which must be an error of some sort.

Changing the data types of the price and odometer attributes, and cleaning and translating the values to English:

```python
autos["price"] = (autos["price"].str.replace('$', '')
                  .str.replace(',', '').astype(int)
                 )
```

```python
autos["odometer"] = (autos["odometer"].str.replace('km', '')
                    .str.replace(',', '')
                     .astype(int)
                    )
```

```python
autos["odometer_km"] = autos["odometer"]
```

```python
autos.drop("odometer", axis=1, inplace=True)
autos.drop("seller", axis=1, inplace=True)
autos.drop("offer_type", axis=1, inplace=True)
autos.drop("num_of_pics", axis=1, inplace=True)
```

```python
autos
```

```python
autos["vehicle_type"] = (autos["vehicle_type"].str.replace('limousine', 'sedan')
                         .str.replace('kleinwagen', 'small car')
                         .str.replace('kombi', 'station wagon')
                         .str.replace('cabrio', 'convertible')
                        )
```

```python
autos["gearbox"] = (autos["gearbox"].str.replace('manuell', 'manual')
                   .str.replace('automatik', 'automatic')
                   )
```

```python
autos["model"] = autos["model"].str.replace('andere', 'other')
```

```python
autos["fuel_type"] = autos["fuel_type"].str.replace('benzin', 'gas')
```

```python
autos["unrepaired_damage"] = autos["unrepaired_damage"].str.replace('nein', 'no').str.replace('ja', 'yes')
```

I wanted to look at the really high price values in this dataset.

I chose to examine above 250,000 in price because leading up to this point when ascending in price value, there are a fair number of values relatively evenly spaced - i.e., they go up in increments of 1k, 5k, sometimes a couple dozen k at the high end. After 350k though, the next value is 999,990 - almost 3x the last value. It gets even more extreme from there.   I'm figuring some of these might be classic cars, luxury vehicles, custom jobs, things people have babied. They could also be random price inputs.

This turned out to be correct.  There are some Porsches, Ferraris, and Rolls-Royces in the 200s, but most of the values above 350,000 seem to be randomly entered values - things like 12345678 or 11111111.  The vehicles associated with them weren't luxury vehicles or classic cars, so it's safe to drop them rather than changing the price on them, because there are only 14 of them which is less than 0.03% of the total dataset.

Conversely, I'm not removing any low values. I'm interpreting these values as asking prices rather than what was actually received. Setting a low starting price, even if it's 0, also isn't unheard of. There are a lot of very low values in this data and excluding them would have meant excluding a lot of the data, which would make for a weaker set of conclusions.  A whole 25 percent of the dataset is below around $1000, which is very cheap for a vehicle.  In this instance, I believe it makes sense to leave these low values in.

Removing the values from before 1911 and after 2016 (the year the data was collected) brought the mean registration year down to 2002 while the median stayed at 2003.  The two values were brought closer with the removal of these outliers.

The last_seen column tells us that not only are there more postings created in the early spring, but that postings are also most likely to be taken offline in March or early April. Although ads are frequently placed throughout the year, from this data we can surmise that the crawler usually sees them last in the spring, since it seems to crawl on at least a daily basis.  Presumably they are sold and taken offline at this point.

```python
hot_brands_counts = autos["brand"].value_counts().head(20)
```

```python
hot_brands = hot_brands_counts.index
```

Next, I've chosen to aggregate by the 20 most frequently listed brands.

```python
price_aggregate = {}
for name in hot_brands:
    price_aggregate[name] = autos.loc[autos["brand"] == name, "price"].mean()
price_aggregate
```

The cars in this list with the highest average selling price are the Mini (average 10460), Audi (average 9903), and Mercedes Benz (average 8485).  The next highest is over 2000 less - the Skoda at 6334.  Other cars ("sonstige_autos") actually account for the total highest average prices, at 10,909.  I wanted to see what items were actually in this set:

```python
mileage_aggregate = {}
for name in hot_brands:
    mileage_aggregate[name] = autos.loc[autos["brand"] == name, "odometer_km"].mean()
mileage_aggregate
```

```python
price_aggregate = {}
for name in hot_brands:
    price_aggregate[name] = autos.loc[autos["brand"] == name, "price"].mean()
price_aggregate
```

```python
mileage_series = pd.Series(mileage_aggregate)
price_series = pd.Series(price_aggregate)
```

```python
price_mileage_by_brand = pd.DataFrame(mileage_series, columns=['mileage'])
price_mileage_by_brand
```

```python
price_mileage_by_brand["price"] = price_series
```

```python
price_mileage_by_brand.sort_values(by='mileage', ascending=False)
```

The luxury vehicles towards the top of this list have among the highest mileage as well as the highest prices - BMW, Mercedes, and Audi are in second, third, and fourth place respectively.  The other cars at the top of the average mileage list cost significantly less.  This suggests that the luxury cars are still worth more despite their relatively high mileage, at least in the eyes of the owners (if these prices are indeed asking prices rather than selling prices).

Interestingly, the sonstige_autos are at the bottom of the list in terms of average mileage, despite having the highest average prices.  The Mini is very similar to the sonstige_autos in both price and mileage, differing in less than 500 for each value.

The data for the mileage seems to be bracketed rather than precise amounts.  The values are all rounded to the nearest thousand, with 150,000km being the most frequent and also the highest value.  This lack of precision coupled with the fact that the price listings may only be asking prices, means that this is very much an estimate in terms of price-mileage correlation.

```python
model_counts = {}
for model in autos["model"]:
    if model not in model_counts:
        model_counts[model] = 0
    model_counts[model]+=1
sorted_model_counts = sorted(model_counts, key=lambda x: model_counts[x], reverse=True)
for k in sorted_model_counts:
    print("{} : {}".format(k, model_counts[k]))
```

The top 12 most common models in this data set are Deutsche - VW, BMW, Audi, and Mercedes Benz.  This is not including the "other" and "nan" values.  The Ford Focus and Fiesta are the next most common on the list.  The VW Golf, which is the top most common model, has almost half again as many listings as the runner up, which is the BMW 3 series.
